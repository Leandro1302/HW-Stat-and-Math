```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
```

```python
n = 5

#Test problem
A = np.random.randn(n, n)
x_true = np.ones((n,))
y = A @ x_true
```

```python
cond_numb_2 = np.linalg.cond(A, p=2)
print(f"Condition number with 2-norm: {cond_numb_2}")
```

```python
cond_numb_inf = np.linalg.cond(A, p=np.inf)
print(f"Condition number with inf-norm: {cond_numb_inf}")
```

### We can notice that there are no significant differences between the conditioning number calculated in 2-norm and inf-norm

```python
x_sol = np.linalg.solve(A, y)
x_sol
```

```python
E_rel = np.linalg.norm(x_true - x_sol, 2) / np.linalg.norm(x_true, 2)
print(f"Relative error: {E_rel}")
```

```python
rel_errors = []
for n in range(10,110,10):
    print(f'n: {n}')
    A = np.random.randn(n, n)
    x_true = np.ones((n,))
    y = A @ x_true
    
    x_sol = np.linalg.solve(A, y)
    E_rel = np.linalg.norm(x_true - x_sol, 2) / np.linalg.norm(x_true, 2)
    rel_errors.append(E_rel)
    
    print(f'The relative error is: {E_rel}')
    print('')

plt.figure(figsize=(20,10))
plt.title('Relative error')
plt.plot(range(10,110,10), rel_errors)
plt.scatter(range(10,110,10),rel_errors)
plt.grid()
plt.xlabel(r"$n$")
plt.ylabel(r"$E_rel$")

plt.show()
```

```python
condition_numbers_2norm = []
condition_numbers_infnorm = []
rel_errors = []
for n in range(10,110,10):
    print(f'n: {n}')
    A = np.random.randn(n, n)
    x_true = np.ones((n,))
    y = A @ x_true
    
    
    cond_numb_2 = np.linalg.cond(A, p=2)
    cond_numb_inf = np.linalg.cond(A, p=np.inf)
    condition_numbers_2norm.append(cond_numb_2)
    condition_numbers_infnorm.append(cond_numb_inf)
    
    
    print(f'Condition number in 2-norm of the random matrix A is {cond_numb_2}')
    print(f'Condition number in inf-norm of the random matrix A is {cond_numb_inf}')
    
    x_sol = np.linalg.solve(A, y)
    E_rel = np.linalg.norm(x_true - x_sol, 2) / np.linalg.norm(x_true, 2)
    rel_errors.append(E_rel)
    
    print(f'The relative error is: {E_rel}')
    print('')

plt.figure(figsize=(20, 5))
plt.title('Condition number')
plt.plot(range(10,110,10), condition_numbers_2norm, color = 'red', label = "2-norm")
plt.plot(range(10,110,10), condition_numbers_infnorm, color = 'y', label = "inf-norm")
plt.grid()
plt.legend()
plt.xlabel(r"$n$")
plt.ylabel(r"$k_2(A)$")

```

```python
condition_numbers_2norm = []
condition_numbers_infnorm = []
nsize = [5,10,15,20,25,30]
rel_errors = []
for n in nsize:
    v = list(range(1,n+1))
    print(f'n: {n}')
    A = np.vander(v, n)
    x_true = np.ones((n,))
    y = A @ x_true
    
    cond_numb_2 = np.linalg.cond(A, p=2)
    cond_numb_inf = np.linalg.cond(A, p=np.inf)
    condition_numbers_2norm.append(cond_numb_2)
    condition_numbers_infnorm.append(cond_numb_inf)
    
    print(f'Condition number in 2-norm of the Vandermonde matrix A is {cond_numb_2}')
    print(f'Condition number in inf-norm of the Vandermonde matrix A is {cond_numb_inf}')
    
    x_sol = np.linalg.solve(A, y)
    E_rel = np.linalg.norm(x_true - x_sol, 2) / np.linalg.norm(x_true, 2)
    rel_errors.append(E_rel)
    
    print(f'The relative error is: {E_rel}')
    print('')


plt.figure(figsize=(20, 10))
plt.suptitle("Condition number and relative errors of Vandermonde Matrix", fontsize=20)
plt.subplot(2,1,1)
plt.title('Condition number')
plt.plot(nsize, condition_numbers_2norm, color = 'red', label = "2-norm")
plt.plot(nsize, condition_numbers_infnorm, color = 'y', label = "inf-norm")



plt.grid()
plt.xlabel(r"$n$")
plt.ylabel(r"$k_2(A)$")
plt.legend()

plt.subplot(2,1,2)

plt.title('Relative error')
plt.plot(nsize, rel_errors)
plt.scatter(nsize,rel_errors)
plt.grid()
plt.xlabel(r"$n$")
plt.ylabel(r"$E_rel$")



plt.show()
```

```python
condition_numbers_2norm = []
condition_numbers_infnorm = []
rel_errors = []
for n in range(4,13):
    print(f'n: {n}')
    A = scipy.linalg.hilbert(n)
    x_true = np.ones((n,))
    y = A @ x_true
    
    cond_numb_2 = np.linalg.cond(A, p=2)
    cond_numb_inf = np.linalg.cond(A, p=np.inf)
    condition_numbers_2norm.append(cond_numb_2)
    condition_numbers_infnorm.append(cond_numb_inf)
    
    print(f'Condition number in 2-norm of the Vandermonde matrix A is {cond_numb_2}')
    print(f'Condition number in inf-norm of the Vandermonde matrix A is {cond_numb_inf}')
    
    x_sol = np.linalg.solve(A, y)
    E_rel = np.linalg.norm(x_true - x_sol, 2) / np.linalg.norm(x_true, 2)
    rel_errors.append(E_rel)
    
    print(f'The relative error is: {E_rel}')
    print('')

plt.figure(figsize=(20, 10))
plt.suptitle("Condition number and relative errors of Hilbert Matrix", fontsize=20)
plt.subplot(2,1,1)
plt.title('Condition number')
plt.plot(range(4,13), condition_numbers_2norm, color = 'red', label = "2-norm")
plt.plot(range(4,13), condition_numbers_infnorm, color = 'y', label = "inf-norm")



plt.grid()
plt.xlabel(r"$n$")
plt.ylabel(r"$k_2(A)$")
plt.legend()

plt.subplot(2,1,2)

plt.title('Relative error')
plt.plot(range(4,13), rel_errors)
plt.scatter(range(4,13),rel_errors)
plt.grid()
plt.xlabel(r"$n$")
plt.ylabel(r"$E_rel$")
```

### We know that both the Hilbert and Vandermonde matrices are ill-conditioned so the relative error increases exponentially as the n increases.

```python
def machine_eps():
    eps = 1.0
    while (1.0 + eps) > 1.0:
        eps /= 2
    return eps

# Test the function
print(f"Machine precision values: {machine_eps()}")
```
